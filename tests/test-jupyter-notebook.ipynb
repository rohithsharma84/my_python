Testing how list slicing works

```python
#Create a list of names
names = ["Alice", "Bob", "Charlie", "David", "Eve"]

# Print the list
for i,name in enumerate(names):
    print("%d - %s" % (i, name))

# Print a slice of the list
print("Slicing the list using indexes 1 to 3: %s" % names[1:3])

# Add more names to the list
names.extend(["Frank", "Grace", "Heidi"])
print("Added more names using extend(): %s" % names)

# Remove one name at a time from the end of the list
print("Using pop() to remove the last element: %s" % names.pop())
print(names)
```

Testing dictionaries

```python
grades = {"Alice": 85, "Bob": 90, "Charlie": 78, "David": 92, "Eve": 88}
print(grades["Alice"])
grades["Alice"] = 40
print(grades["Alice"])
print(grades.keys())
print(grades.values())
print(grades.items())
grades.pop("Alice")
print(grades)
```

Testing how sets work

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print("set1.union(set2) : %s" % (set1.union(set2)))
print("set1 | set2 : %s" % (set1 | set2))
print("set1.intersection(set2) : %s" % (set1.intersection(set2)))
print("set1 & set2 : %s" % (set1 & set2))
print("set1.difference(set2) : %s" % (set1.difference(set2)))

# Add a new element to a set
set1.add(6)
print(set1)

# Trying to add a duplicate element to a set.. it does not change the set
set1.add(3)
print(set1) 
```

for-elif-else loops

```python
name="RohitSharma"
for i in name:
    if i == 'a':
        continue # Skip the rest of the loop and iterate if char is 'a'
    elif i == 't':
        break # Exit the loop if char is 't'
    else:
        print(i)
```

Create a simple function that check the age category given an integer as input

```python
def age_group(age):
    if age < 13:
        return "Child"
    elif age < 20:
        return "Teenager"
    elif age < 65:
        return "Adult"
    else:
        return "Senior"

print(age_group(90))
```

List comprehension example

```python
# Get a list of squares of even numbers using list comprehension
evens = [x**2 for x in range(1,10) if x % 2 == 0]
print(evens)

# Traditional way to create a list of squares of even numbers
evens =[]
for x in range(1,10):
    if x % 2 == 0:
        evens.append(x**2) 
print(evens)
```
